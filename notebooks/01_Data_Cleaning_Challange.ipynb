```python
import pandas as pd
```

```python
# code to ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
database = pd.read_csv('../001_Input_files/01_Data-rawdata.csv')
database.head()
```

```python
# remove all rows with missing values in Row Type column and remove columns name%%!d Unnamed: 9 and Unnamed: 10
database = database[database['Row Type'].notna()]
database = database.drop(['Unnamed: 9', 'Unnamed: 10'], axis=1)
database.head()
```

```python
working_database = database.copy()
working_database.head()
```

```python
column_values = []
counter = 0
for i in working_database['Row Type']:
    if 'first name' in i:
        counter += 1
    column_values.append(counter)
    
```

```python
Iter_cols = working_database
Iter_cols['Iteration'] = column_values
```

```python
Iter_cols
```

```python
drop_extra_rows = Iter_cols[Iter_cols['Row Type'] != 'Row Type']
drop_extra_rows
```

```python
# df[df['class'].str.contains('i', na=False)]
name_dataframe = drop_extra_rows[drop_extra_rows['Row Type'].str.contains('first name', na=False)]

# dropping extra column_values
name_dataframe.drop(columns=['Speed1', 'Speed2', 'Electricity','Effort', 'Weight', 'Torque'], 
                    inplace=True)

# renaming the column_values
name_dataframe.rename(columns={'Row Type': 'First Name',
                               'Iter Number' : 'Last Name',
                               'Power1': 'Date'}, inplace=True)

name_dataframe['First Name'] = name_dataframe['First Name'].str[12:]
name_dataframe['Last Name'] = name_dataframe['Last Name'].str[11:]
name_dataframe['Date'] = name_dataframe['Date'].str[6:]

name_dataframe
```

```python
no_name_dataframe = drop_extra_rows[drop_extra_rows['Row Type'] != 'first name: Person']
no_name_dataframe
```

```python
output_df = pd.merge(left=name_dataframe, right=no_name_dataframe, how='inner', on='Iteration')
```

```python
output_df
```

```python
# Output File 
output_df.to_csv('../001_Input_files/02_Data-output.csv', index=False)
```
